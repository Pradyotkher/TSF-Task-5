```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('C:\\Users\\Pradyot\\Desktop\\matches.csv')
df=df.drop(['date'],axis=1)
df
```

```python
df.info()
# We have some null values in the following fields City,Winner,player_of_match,umpire1,umpire2,umpire3
```

```python
# Filling Null Values with 0
df['city'] = df['city'].fillna(0)
df['winner'] = df['winner'].fillna(0)
df['player_of_match'] = df['player_of_match'].fillna(0)
df['umpire1'] = df['umpire1'].fillna(0)
df['umpire2'] = df['umpire2'].fillna(0)
df['umpire3'] = df['umpire3'].fillna(0)
df.info()
```

```python
df
```

```python
winner = df['winner']
toss = df['toss_decision']
# percent of teams who won the match after winning the toss 
# percentage of teams winning after batting first and fielding first . 
winner_based_on_toss = {'Winner':winner,'Toss_decision':toss}
data = pd.DataFrame(winner_based_on_toss)
count=0
for i in data['Toss_decision']:
    if i=='field':
         count=count+1
percent = count/755
per = percent*100
print(per,'% of teams won the game if the chose to field first')
won = 100-per
print(won,'% of Teams won the game by choosing to Bat first')
```

```python
sns.factorplot('Toss_decision',data=data,kind='count')
# The graph shows the number of teams won the match when decided to field and bat first respectively 
```
